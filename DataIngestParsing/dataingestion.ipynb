TextLoader -> read single File

```python
from langchain_community.document_loaders import TextLoader


loader = TextLoader("data/text_files/ospython.txt",encoding='utf-8')
documents = loader.load()

# type(loader)
print(type(documents))
print(documents)
```

DirectoryLoader -> multiple Text Files

```python
from langchain_community.document_loaders import DirectoryLoader

dirloader = DirectoryLoader(
    "data/text_files", 
    glob= "*.txt", 
    loader_cls = TextLoader, 
    loader_kwargs = {"encoding" : 'utf-8'}, 
    show_progress = True
)

documents = dirloader.load()

print(type(documents))
print(documents)

print(f"Documents: {len(documents)} documents \n")
for i,doc in enumerate(documents):
    print(f"Document {i+1}")
    print(f"source = {doc.metadata['source']}")
    print(f"length = {len(doc.page_content)}\n")
```

```python
from langchain_text_splitters import (
    RecursiveCharacterTextSplitter,
    CharacterTextSplitter,
    TokenTextSplitter
)
print(documents)
```

```python
text = documents[0].page_content
print(text)
```

Character Text Splitter

```python
char_splitter = CharacterTextSplitter(
    separator = ' ',
    chunk_size = 200,
    chunk_overlap = 20,
    length_function = len
)

char_chunks = char_splitter.split_text(text)

print(f"creates {len(char_chunks)} chunks")
print(f"chunk[0] = {char_chunks[0][:100]}")
```

Recurssive Text Splitter


Most used 

```python
recur_splitter = RecursiveCharacterTextSplitter(
    separators = ['\n\n','\n'," ",""],
    chunk_size = 200,
    chunk_overlap = 20,
    length_function = len
)
rec_chunks = recur_splitter.split_text(text)

print(f"creates {len(rec_chunks)} chunks")
print(f"chunk[0] = {rec_chunks[0][:100]}")
```

```python
from langchain_community.document_loaders import(
    PyPDFLoader,
    PyMuPDFLoader
)
```

PyPDFLoader is the most recommended pdf loader

```python
print("PyPDFLoader \n")

try:
    pypdf_loader = PyPDFLoader("data/pdfs/Midterm_Solution.pdf")
    pypdf_docs = pypdf_loader.load()
    print(pypdf_docs)
    print(f"total pages is {len(pypdf_docs)}")
    print(f"Page 1 is {pypdf_docs[1].page_content}")

except Exception as e:
    print(f"Error is {e}") 
    
```

PyPDFLoader is more useful when the pdf is standard texts pdf
