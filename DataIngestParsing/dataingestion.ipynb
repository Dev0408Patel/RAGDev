```python
char_splitter = CharacterTextSplitter(
    separator = ' ',
    chunk_size = 200,
    chunk_overlap = 20,
    length_function = len
)

char_chunks = char_splitter.split_text(text)

print(f"creates {len(char_chunks)} chunks")
print(f"chunk = {char_chunks}")
print(f"chunk[0] = {char_chunks[0][:100]}")
```

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_core.documents import Document
```

```python
class SmartPDFProcessor:
    """Advanced pdf processing with error handling"""
    def __init__(self,chunk_size=1000,chunk_overlap=30):
        self.chunk_size = chunk_size
        self.chunk_overlap = chunk_overlap
        self.text_splitter = RecursiveCharacterTextSplitter(
            chunk_size = chunk_size,
            chunk_overlap = chunk_overlap,
            separators = [" "]
        )

    def process_pdf(self,pdf_path : str) -> List[Document]:
        """process pdf with smart chunking process Data"""
        loader = PyPDFLoader(pdf_path)
        pages = loader.load()

        processed_chunks = []
        for ind,page in enumerate(pages):
            cleaned_text = self.clean_text(page.page_content)

            if len(cleaned_text) < 50 :
                continue

            chunks = self.text_splitter.create_documents(
                texts = [cleaned_text],
                metadatas = [{
                **page.metadata,
                "page" : ind + 1,
                "total_pages": len(pages),
                "chunking_method" : "smart_pdf_processor",
                "char_count": len(cleaned_text)
                }]
                
            )
            processed_chunks.append(chunks)

        return processed_chunks

    def clean_text(self, text:str) -> str:
        text = " ".join(text.split())
        return text



```

```python
preprocessor = SmartPDFProcessor()
```

```python
try:
    smart_chunks = preprocessor.process_pdf("data/pdfs/Midterm_Solution.pdf")
    print(f"processed into {(smart_chunks)} chunks")
    if smart_chunks:
        for ind,chunks in enumerate(smart_chunks):
            print(f"Chunk : {ind+1}")
            print()
    

except Exception as e:
    print(f"Error caught while loading : {e}")
```
